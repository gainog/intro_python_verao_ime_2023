
5) Crie um objeto chamado **criancas_cherbourg** que seleciona as crianças com menos de 12 anos E que embarcaram na cidade de Cherbourg.


10) Agrupe pela classe e conte a quantidade de pessoas por classe


```python
df.groupby("classe").count()

# Classe 1 = 216, 2 = 184, 3 = 491
```


11) Agrupe pela classe e pela coluna sobreviveu e conte a quantidade de pessoas que sobreviveu de acordo com cada classe. Quem sobreviveu mais? A classe mais alta ou a mais baixa?


```python
my_analysis = df[["classe","sobreviveu","id_passageiro"]]
my_count = my_analysis.groupby(["classe","sobreviveu"]).count()
my_count = my_count.rename(columns={"id_passageiro":'count'})
my_count["total_classe"] = my_count.groupby(["classe"]).transform(func=sum)
my_count["percentual"] = round(my_count["count"] / my_count["total_classe"] * 100,2)
my_count

# Resp: Tanto em número absoluto quanto em percentual relativo à classe,
#       a classe 1 tem maior número de sobreviventes
```


12) Agrupe pelo sexo e pela coluna sobreviveu e conte a quantidade de pessoas que sobreviveram de acordo com o sexo. Quem sobreviveu mais? Homens ou mulheres?

```python
my_analysis = df[["sexo","sobreviveu","id_passageiro"]]
my_count = my_analysis.groupby(["sexo","sobreviveu"]).count()
my_count = my_count.rename(columns={"id_passageiro":'count'})
my_count["total_sexo"] = my_count.groupby(["sexo"]).transform(func=sum)
my_count["percentual"] = round(my_count["count"] / my_count["total_sexo"] * 100,2)
my_count

# Resp: Tanto em número absoluto quanto em percentual relativo ao sexo,
#       mais mulheres sobreviveram
```
